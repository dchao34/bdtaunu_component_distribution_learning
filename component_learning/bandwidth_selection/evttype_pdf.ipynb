```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from kde_visual_utils import plot_data2d, plot_contour2d
```

```python
evttype = 1

scatter_csv_fname = 'visual_output/evttype{0}_scatter.csv'.format(evttype)
kde2d_csv_fname = 'visual_output/evttype{0}_kde.csv'.format(evttype)
akde2d_csv_fname = 'visual_output/evttype{0}_akde.csv'.format(evttype)

gridsearch_csv_fname = 'visual_output/evttype{0}_gridsearch.csv'.format(evttype)
gridsearch_pdf_fname = 'evttype{0}_gridsearch.pdf'.format(evttype)
```

```python
fig = plt.figure(figsize=(24,24))

ax = fig.add_subplot(2,2,1)
plot_data2d(scatter_csv_fname, ax=ax, title=r'input sample', xlim=[-0.005,0.11], ylim=[-0.05,1.05])
ax = fig.add_subplot(2,2,3)
plot_contour2d(kde2d_csv_fname, ax=ax, title=r'cross validated', 
               nlevels=10, colorbar=False, contour_fmt='%1.0f')
ax = fig.add_subplot(2,2,4)
plot_contour2d(akde2d_csv_fname, ax=ax, title=r'cross validated and adapted', 
               nlevels=10, colorbar=False, contour_fmt='%1.0f')
```

```python
fig = plt.figure(figsize=(10,8))

ax = fig.add_subplot(1,1,1)
plot_contour2d(gridsearch_csv_fname, ax=ax, 
               title=r'least squares cross validation score', 
               xlabel=r'$h_1$', ylabel=r'$h_2$',
               nlevels=10, nlevels_f=100, colorbar=True, 
               cmap=plt.cm.YlOrRd, 
               contour_fmt='%1.2f')

fig.savefig(gridsearch_pdf_fname)
```
