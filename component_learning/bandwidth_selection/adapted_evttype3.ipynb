```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from kde_visual_utils import plot_data2d, plot_contour2d
```

```python
fig = plt.figure(figsize=(24,24))

ax = fig.add_subplot(2,2,1)
plot_data2d('adapted_evttype3_scatter.csv', ax=ax, title=r'input sample', xlim=[0.0,0.1], ylim=[0,1.0])
ax = fig.add_subplot(2,2,3)
plot_contour2d('adapted_evttype3_kde.csv', ax=ax, title=r'cross validated', 
               nlevels=10, colorbar=False, contour_fmt='%1.0f')
ax = fig.add_subplot(2,2,4)
plot_contour2d('adapted_evttype3_akde.csv', ax=ax, title=r'cross validated and adapted', 
               nlevels=10, colorbar=False, contour_fmt='%1.0f')
```

```python
fig = plt.figure(figsize=(10,8))

ax = fig.add_subplot(1,1,1)
plot_contour2d('adapted_evttype3_gridsearch.csv', ax=ax, title=r'least squares cross validation score', 
               nlevels=100, nlevels_f=300, colorbar=True, cmap=plt.cm.YlOrRd_r, contour_fmt='%1.2f')
```
