```python
%matplotlib inline
import sys
sys.path.append('../../utils')
import numpy as np
import matplotlib.pyplot as plt
from matplotlib_visualization_utils import plot_contour2d, plot_data2d, plot_data1d, plot_data1d_histogram
from matplotlib_visualization_utils import plot_kde2d_summary
import iopro
```

```python
evttype = 1

data_fname = 'data/evttype{0}.bw.train.csv'.format(evttype)

kde2d_csv_fname = 'visual_output/evttype{0}.kde2d.csv'.format(evttype)
kde2d_pdf_fname = 'evttype{0}_kde2d.pdf'.format(evttype)

kde1dx_csv_fname = 'visual_output/evttype{0}.kde1d_x.csv'.format(evttype)
kde1dx_pdf_fname = 'evttype{0}_kde1dx.pdf'.format(evttype)

kde1dy_csv_fname = 'visual_output/evttype{0}.kde1d_y.csv'.format(evttype)
kde1dy_pdf_fname = 'evttype{0}_kde1dy.pdf'.format(evttype)

kde_combined_pdf_fname = 'evttype{0}_combined.pdf'.format(evttype)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
plot_contour2d(kde2d_csv_fname, ax=ax, title=r'kernel density', nlevels=10, colorbar=False)
fig.savefig(kde2d_pdf_fname)
```

```python
arr = iopro.genfromtxt(data_fname)

fig = plt.figure(figsize=(6*1.618,6))
ax = fig.add_subplot(1,1,1)
plot_data1d(kde1dx_csv_fname, ax=ax, title=r'marginal density in X1', xlabel=r'$X_1$')
plot_data1d_histogram(arr[:,0], weights=arr[:,2],bins=100,normed=True);
fig.savefig(kde1dx_pdf_fname)
```

```python
arr = iopro.genfromtxt(data_fname)

fig = plt.figure(figsize=(6*1.618,6))
ax = fig.add_subplot(1,1,1)
plot_data1d(kde1dy_csv_fname, ax=ax, title=r'marginal density in X2', xlabel=r'$X_2$')
plot_data1d_histogram(arr[:,1], weights=arr[:,2],bins=100,normed=True);
fig.savefig(kde1dy_pdf_fname)
```

```python
fig, ax1, ax2, ax3 = plot_kde2d_summary(data_fname, kde2d_csv_fname,
                                        kde1dx_csv_fname, kde1dy_csv_fname, 
                                        title='Event Type {0}'.format(evttype), 
                                        draw_scatter=False,
                                        scatter_undersample=0.1, scatter_s=10,
                                        xlim=[-0.005, 0.11], ylim=[-0.05,1.05]);
fig.savefig(kde_combined_pdf_fname)
```
