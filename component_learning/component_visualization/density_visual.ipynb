```python
%matplotlib inline
import sys
sys.path.append('../../utils')
import numpy as np
import matplotlib.pyplot as plt
from matplotlib_visualization_utils import plot_contour2d, plot_data2d, plot_data1d, plot_data1d_histogram
from matplotlib_visualization_utils import plot_kde2d_summary
import iopro
```

```python
fig = plt.figure(figsize=(24,12))
ax = fig.add_subplot(1,2,1)
plot_data2d('visual_output/evttype1.scatter.csv', ax=ax, title=r'input sample', marker='.',
            xlim=[-0.005,0.11], ylim=[-0.05,1.05])
ax = fig.add_subplot(1,2,2)
plot_contour2d('visual_output/evttype1.kde2d.csv', ax=ax, title=r'kernel density', nlevels=10, colorbar=False)
```

```python
arr = iopro.genfromtxt('data/evttype1.bw.train.csv')

fig = plt.figure(figsize=(12*1.618,6))
ax = fig.add_subplot(1,2,1)
plot_data1d('visual_output/evttype1.kde1d_x.csv', ax=ax, title=r'marginal density in X1', xlabel=r'$X_1$')
plot_data1d_histogram(arr[:,0], weights=arr[:,2],bins=100,normed=True);
ax = fig.add_subplot(1,2,2)
plot_data1d('visual_output/evttype1.kde1d_y.csv', ax=ax, title=r'marginal density in X2', xlabel=r'$X_2$')
plot_data1d_histogram(arr[:,1], weights=arr[:,2],bins=100,normed=True);
```
